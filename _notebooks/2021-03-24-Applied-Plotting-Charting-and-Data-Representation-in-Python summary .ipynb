## Line plots

```python
# Data for plotting
t = np.arange(0, 2, 0.01)
s = 1 + np.sin(2 * np.pi * t)

# plt.subplots()해당하는 fig와 axes들 반환
fig,ax = plt.subplots()

ax.plot(t,s)
ax.plot(t,t)
# ax.set - ax에 해당하는 axes의 속성 지정 가능,,
ax.set(xlabel='time (s)', ylabel='voltage (mV)',
      title='About as simple as it gets, folks')
# grid 생성,ax.grid(b=None, which='major', axis='both', **kwargs)
ax.grid()
# 선사이 색체우기
# plt.gca().fill_between(x축, 선1,선2,facecolor=채울색,alpha=색 진하기)
plt.gca().fill_between(t,s,t,
                      facecolor='red',
                      alpha=0.5)
plt.show()
```

```python
# 날짜데이터를 이용한 시각화

plt.figure()

observation_dates = np.arange('2017-01-01','2017-01-09',dtype = 'datetime64[D]')
# observation_dates

linear_data = np.array([1,2,3,4,5,6,7,8])
# linear_data
plt.plot(observation_dates,linear_data,'-o')
plt.plot(observation_dates,linear_data**2,'-o')
# plt.xticks(rotation=45,size=4)
plt.xticks(rotation=45, size=5)
plt.xlabel('date')
plt.ylabel('Units')
plt.title('expnetial vs. linear performances')
```

## Bar Charts

## Histograms
- plt.hist(bins=계급수)

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,sharex=True)
axs = [ax1,ax2,ax3,ax4]

for n in range(len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size = sample_size)
    axs[n].hist(sample)
    axs[n].set_title('n={}'.format(sample_size))
```

## Animatinos

```python
import matplotlib.animation as animation

n=100
#n개의 숫자를 무작위로 생성
x= np.random.randn(n)

# a-b 사이의 숫자중 c개의 정수를 무작위로 생성해라
# np.random.randint(a,b,c)
```

```python
def update(curr):
    if curr==n:
        a.event_source.stop()
    plt.cla() # clear axes
    bins = np.arange(-4,4,0.5)
    plt.hist(x[:curr],bins=bins)
    plt.axis([-4,4,0,30])
    plt.gca().set_title('Sampling the Normal Distiribution') #현재 figure의 제목 지정
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_xlabel('Value')
    plt.annotate('n={}'.format(curr),[3,27])
    #(3,27)위치에 'n={}'.format(curr)가 표시됨
    
```

```python
fig = plt.figure()
a = animation.FuncAnimation(fig, update, interval=100)
# fig에 update를 계속 불러오면서 interval만큼 실행
```

## Interactivity

```python
plt.figure()
data = np.random.randn(10)
plt.plot(data)

def onclick(event):
    plt.cla()
    plt.plot(data)
    plt.gca().set_title('Event at pixels {},{} \n and {},{}'.format(event.x,event.y,event.xdata,event.ydata))
    
plt.gcf().canvas.mpl_connect('button_press_event', onclick)
```

```python
# gca - get current axes
# gcf - get current figure
# cla - clear current axes
# clf - clear current figure
```

```python
from random import shuffle
origins = ['China','Brazil','India','USA','Canada','UK','Iraq',
          'Germany','Chile','Mexico']
# origins 섞기
shuffle(origins)

df = pd.DataFrame({'height':np.random.rand(10),
                  'weight':np.random.rand(10),
                  'origin':origins})
```

```python
df
```

```python
plt.figure()
plt.scatter(df['height'], df['weight'],picker=5)
plt.gca().set_ylabel('Weight')
plt.gca().set_xlabel('Height')
```

```python
def onpick(event):
    origin = df.iloc[event.ind[0]]['origin']
    plt.gca().set_title('Selected item came from {}'.format(origin))

# tell mpl_connect we want to pass a 'pick_event' into onpick when the event is detected
plt.gcf().canvas.mpl_connect('pick_event', onpick)
#클릭할때마다 점에 해당하는 나라이름 출력
```

plt.plot()의 kind parameter가 아닌 `DataFrame.plot.kind`로 그래프 종류 선택가능.

`kind` :
- `'line'` : line plot (default)
- `'bar'` : vertical bar plot
- `'barh'` : horizontal bar plot
- `'hist'` : histogram
- `'box'` : boxplot
- `'kde'` : Kernel Density Estimation plot
- `'density'` : same as 'kde'
- `'area'` : area plot
- `'pie'` : pie plot
- `'scatter'` : scatter plot
- `'hexbin'` : hexbin plot

## Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook
```

```python
np.random.seed(1234)

v1 = pd.Series(np.random.normal(0,10,1000),name='v1')
v2 = pd.Series(2*v1 + np.random.normal(60,15,1000),name='v2')
```

```python
plt.figure()

plt.hist(v1, alpha=0.7, bins=np.arange(-50,150,5), label='v1');
plt.hist(v2, alpha=0.7, bins=np.arange(-50,150,5), label='v2');
plt.legend()
```

```python
# 쌓여있는 barchart에 kernel density estimation그리기

plt.figure()
plt.hist([v1,v2], histtype='barstacked', density=True);
v3 = np.concatenate((v1,v2))
sns.kdeplot(v3);
```

```python
plt.figure()

#위에서 그린 그래프를 그리는 다른 방법
#각 그래프의 매개변수를 각각 지정 가능
sns.distplot(v3, hist_kws={'color':'Teal','width':2}, kde_kws={'color':'Navy'});
```

```python
sns.jointplot(v1,v2,alpha=0.4)
```

```python
sns.jointplot(v1,v2,kind='hex');
#점 모양을 육각형(hexagon)으로 지정
```

```python
sns.set_style('white')
# 이후의 그래프들의 스타일 지정
# {darkgrid, whitegrid, dark, white, ticks} 중 선택가능
sns.jointplot(v1,v2,kind='resid',space=0)
# kind : { "scatter" | "reg" | "resid" | "kde" | "hex" } 선택가능
```
